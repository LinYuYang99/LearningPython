```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a=np.loadtxt(fname="data/inflammation-01.csv",delimiter=',')
figure=plt.figure(figsize=(10,3))
axes1= figure.add_subplot(1,3,1)
axes2= figure.add_subplot(1,3,2)
axes3= figure.add_subplot(1,3,3)

axes1.set_ylabel('acerage')
axes1.plot(a.mean(axis=0))

axes2.set_ylabel('maxium')
axes2.plot(a.max(axis=0))

axes3.set_ylabel('minium')
axes3.plot(a.min(axis=0))

figure.tight_layout()
plt.show()
```
